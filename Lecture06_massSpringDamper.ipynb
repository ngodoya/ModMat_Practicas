### Obtener el modelo para el sistema Masa - Resorte - Amortiguador
<br>
<div align='justify'>A continuación se presenta el esquema de una masa $m$ unida a un amortiguador y a un resorte, que están conectados en paralelo. A su vez, sobre la masa se ejerce una fuerza $F$, que puede o no depender del tiempo: </div> <br>

<img src=https://raw.githubusercontent.com/juhincapiem/ModMat/refs/heads/main/Diagrams/MasaResorteAmortiguador.jpg alt="Alternative text" width="400" />


<div align='justify'> Sigamos los pasos de la siguiente metodología para plantear el modelo: </div> <br>


1. Asumir la dirección de movimiento. Para este caso, se asumen que la masa $m$ empieza a moverse hacia la derecha.
<br>
1. Plantear el Diagrama de Cuerpo Libre. Toda fuerza que vaya en la misma dirección de la dirección de movimiento que se asumió, es positiva. 
<br>
1. Aplicar la segunda Ley de Newton: $\sum F = m\cdot a$.
<br>

La ecuación que se debe obtener es:

$$
    ma = -kx - cv + F
$$

Donde $k$ es la constante de resorte y $c$, en algunos textos se pone como $\beta$, es la constante de amortiguamiento.

### Tipos de comportamiento
<br>
<div align='justify'> Sabemos que existen 3 diferentes respuestas (sobre amortiguado, críticamente amortiguado y sub-amortiguado) de un sistemas mecánica compuesto por masa, resorte y amortiguador cuando es perturbado. Hay distintas formas de anticiparnos a estas respuestas. Revisemos una de ellas: </div> 
<br>

1. Proponer un cambio de variable.
$$
    X_1 = x \, \, \text{y} \, \,  X_2 = v
$$
1. Plantear dos ecauciones de evolución.
<br>
$$
\begin{aligned}
\dot{X}_2 &=-\frac{c}{m} X_2-\frac{k}{m} X_1+F \\
\dot{X}_1 &=X_2
\end{aligned}
$$
1. Presentarlo en forma matricial.
<br>
$$
\left[\begin{array}{l}
\dot{x}_1 \\
\dot{x}_2
\end{array}\right]=\left[\begin{array}{cc}
0 & 1 \\
-\frac{k}{m} & -\frac{c}{m}
\end{array}\right]\left[\begin{array}{l}
x_1 \\
x_2
\end{array}\right]+\left[\begin{array}{l}
0 \\
F
\end{array}\right]
$$
1. Encontrar los autovalores para la matriz de coeficientes.
<br>
$$
|A-I\lambda| = 0
$$
<br>
Donde la matriz de coeficientes es:
$$
A = \left[\begin{array}{cc}
0 & 1 \\
-\frac{k}{m} & -\frac{c}{m}
\end{array}\right]
$$
1. Clasificar el comportamiento:
<br>
$$
\begin{array}{l|l}
\textbf { Tipo de respuesta } & \textbf { Autovalores } \\
\hline \text { Sobreamortiguado } & \text {  Reales y diferentes } \\
\hline \text { Críticamente amortiguado } & \text { Reales e iguales } \\
\hline \text { Subamortiguado } & \text { Complejos }
\end{array}
$$

### Solución analítica para la ecuación homogénea
<br>
<div align='justify'> Como son tres tipos de respuesta, esperamos que hayan tres tipos de soluciones analíticas. Para encontrar la respectiva solución, nos conviene darle una forma "especial" a nuestro modelos de segundo orden:</div> 
<br>

$$
    a +2 \gamma v +\omega^2 x= F
$$

Donde $2 \gamma = \frac{c}{m}$ y $\omega^2 = \frac{k}{m}$

Teniéndolo de esta forma, los autovalores se pueden calcular usando las siguientes expresiones:

$$
\lambda_1=-\gamma+\sqrt{\gamma^2-\omega^2}, \quad \lambda_2=-\gamma-\sqrt{\gamma^2-\omega^2}
$$
<br><br>
#### Sobreamortiguado

$$
x(t)=e^{-\gamma t}\left(c_1 e^{\sqrt{\gamma^2-\omega^2} t}+c_2 e^{-\sqrt{\gamma^2-\omega^2} t}\right)
$$

Los coeficientes se calculan de la siguiente manera:

$$
C_2= \frac{x\prime(0) - \lambda_1}{\lambda_2 - \lambda_1}, \quad C_1=x(0) - C_2
$$

<br><br>
#### Críticamente amortiguado

$$
x(t)=e^{-\gamma t}\left(c_1+c_2 t\right)
$$

Los coeficientes se calculan de la siguiente manera:

$$
C_1= x(0), \quad C_2=x\prime(0) + \gamma C_1
$$

<br><br>
#### Subamortiguado

$$
x(t)=e^{-\gamma t}\left(c_1 \cos \left(\sqrt{\omega^2-\gamma^2} t\right)+c_2 \operatorname{sen} \left(\sqrt{\omega^2-\gamma^2} t\right)\right)
$$

Los coeficientes se calculan de la siguiente manera:

$$
C_1= x(0), \quad C_2= \frac{x\prime(0) + \gamma C_1}{\sqrt{\omega^2 - \gamma^2}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

$$
\frac{d^2 x}{d t^2}+5 \frac{d x}{d t}+4 x=0, \quad x(0)=1, \quad x^{\prime}(0)=1
$$

```python
# Paso 01: identificar los parámetros
#Note que es, autonoma, Homogenea, lineal, de orden 2, grado 1, la posición inicial es 1 y la velocidad inicial es 1
a = 1
b = 5
c = 4

# Paso 02: calcular los autovalores
characteristical_polynomial = np.poly1d([a, b, c]) #a, b, c
roots = np.roots(characteristical_polynomial)
lambda_1, lambda_2 = roots[0], roots[1]
# Paso 03: calcular los coeficientes C1 y C2 a partir de las condiciones iniciales
A = np.array([[1, 1],[-4, -1]])
rhs = np.array([[1], [1]])
# Otra opcion para resolver 
#vect_solution = np.linalg.solve(A, b)
#c_1, c_2 = vect_solution[0], vect_solution[1]
# La que usaba
A_inv = np.linalg.inv(A)
vect_solution = A_inv @ rhs
c_1, c_2 = vect_solution[0], vect_solution[1]
# Paso 04: construir función que representa la solución analítica
def analytic_function(t):
    """Returns analytic function"""
    return c_1 * np.exp(lambda_1 * t) + c_2 * np.exp(lambda_2 * t)

# Paso 05: crear vector de tiempo
t0 = 0
tf = 10
dt = 0.02
vect_time = np.arange(t0, tf + dt, dt)

# Paso 06: Inicializar el vector para recibir los valores para la posición
overdamped_analytic_solution = analytic_function(vect_time)

# Paso 07: solucionar
overdamped_analytic_solution = analytic_function(vect_time)

# Paso 08: Gráficar
fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi=100)
ax.plot(vect_time, overdamped_analytic_solution, 'k-.', label="Solución Análitica")

# Carpinteria
ax.grid()
ax.set_title(r"Función solución: $x(t) = -\dfrac{2}{3}e^{-4t} + \dfrac{5}{3}e^{-t}$")
ax.legend()
ax.set_xlim([0, tf])
ax.set_ylim(0, None)
```

```python
# BUSQUEMOS UNA SOLUCIÓN NÚMERICA
# Paso 01: Encuentre el sistema linea 2x2 referente a la EDO de orden 2
# x_1' = x_2 
# x_2' = -5x_2 -4x_1
# Paso 02: crear vector de tiempo
t0 = 0
tf = 10
dt = 0.05
vect_analytical_time = np.arange(t0, tf + dt, dt)
# Paso 03: Euler hacía adelante
def aceleration(x, v):
    """Returns numerical function x(t), using Euler Forward"""
    return -5 * v -4 * x
# Paso 04: Defina los vectores de posición y velocidad
numerical_solution = np.zeros(len(vect_analytical_time))
velocity = np.zeros(len(vect_analytical_time))
# Paso 05: Utilice las condiciones iniciales x(0) = x'(0) = 1 
numerical_solution[0] = 1
velocity[0] = 1

# Paso 06: Ejecutar el ciclo for para el euler Forward
for i in range(len(vect_analytical_time) - 1):
    velocity[i+1] = velocity[i] + aceleration(numerical_solution[i], velocity[i]) * dt
    numerical_solution[i+1] = numerical_solution[i] + velocity[i] * dt

# Paso 07: Gráficar
fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi=100)
ax.plot(vect_time, overdamped_analytic_solution, 'g--', label="Solución Análitica")
ax.plot(vect_analytical_time, numerical_solution, 'y-.', label="Solución Númerica")

# Carpinteria
ax.grid()
ax.set_title("Comparación Solución análitica y Solución Análitica")
ax.legend()
ax.set_xlim([0, tf])
ax.set_ylim(0, None)
```

$$
\frac{d^2 x}{d t^2}+8 \frac{d x}{d t}+16 x=0, \quad x(0)=1, \quad x^{\prime}(0)=1
$$

```python
# Paso 01: identificar los parámetros
a = 1
gamma = 8
omega = 16

# Paso 02: calcular los autovalores
characteristical_polynomial = np.poly1d([a, gamma, omega])
roots = np.roots(characteristical_polynomial)
p1 = roots[0]
p2 = roots[1]
print("P1 = {:.2f} y P2 = {:.2f}".format(p1,p2))
#  Paso Extra, función para crear el sistema lineal:
x_1 = lambda t: np.exp(p1 * t)  
x_2 = lambda t: t * np.exp(p1 * t)
x_1_derivative = lambda t: p1  * np.exp(p1 * t) 
x_2_derivative = lambda t: np.exp(p1 * t) + t * p1 * np.exp(p1 * t) 

# Paso 03: calcular los coeficientes C1 y C2 a partir de las condiciones iniciales
A = np.array([[x_1(0), x_2(0)],[x_1_derivative(0), x_2_derivative(0)]])
b = np.array([[1], [1]])
x_solution = np.linalg.solve(A, b)
C1 = x_solution[0][0]
C2 = x_solution[1][0]

print("C1 = {:.2f} y C2 = {:.2f}".format(C1,C2))

# Paso 04: construir función que representa la solución analítica
def critically_damped(lambda_0, c_1, c_2, t):
    """Returns analytic solution x(t)"""
    return np.exp(lambda_0 * t) * (c_1 + t * c_2 )
# Paso 05: crear vector de tiempo
t0 = 0
tf = 10
dt = 0.02
vect_time_critical = np.arange(t0, tf + dt, dt)
# Paso 06: solucionar
critically_damped_analytic_function = critically_damped(p1, C1, C2, vect_time_critical)

# Paso 07: Graficar y carpinteria
fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi=100)
ax.plot(vect_time_critical, critically_damped_analytic_function, 'g--', label="Solución Análitica")
ax.grid()
ax.set_title("Sistema Críticamente amortiguado")
ax.legend()
ax.set_xlim([0, tf])
ax.set_ylim(0, None)
```

```python
# BUSQUEMOS UNA SOLUCIÓN NÚMERICA
# Paso 01: Encuentre el sistema linea 2x2 referente a la EDO de orden 2
# Paso 02: crear vector de tiempo
t0 = 0
tf = 10
dt = 0.02
vect_analytical_time = np.arange(t0, tf + dt, dt)
# Paso 03: Euler hacía adelante
def aceleration(x, v):
    """Returns numerical function x(t), using Euler Forward"""
    return -8 * v -16 * x
# Paso 04: Defina los vectores de posición y velocidad
numerical_solution = np.zeros(len(vect_analytical_time))
velocity = np.zeros(len(vect_analytical_time))
# Paso 05: Utilice las condiciones iniciales x(0) = x'(0) = 1 
numerical_solution[0] = 1
velocity[0] = 1

# Paso 06: Ejecutar el ciclo for para el euler Forward
for i in range(len(vect_analytical_time) - 1):
    velocity[i+1] = velocity[i] + aceleration(numerical_solution[i], velocity[i]) * dt
    numerical_solution[i+1] = numerical_solution[i] + velocity[i] * dt

# Paso 07: Gráficar
fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi=100)
ax.plot(vect_time_critical, critically_damped_analytic_function, 'g--', label="Solución Análitica")
ax.plot(vect_analytical_time, numerical_solution, 'y-.', label="Solución Númerica")

# Carpinteria
ax.grid()
ax.set_title("Comparación Solución análitica y Solución Análitica")
ax.legend()
ax.set_xlim([0, tf])
ax.set_ylim(0, None)
```


$$

```python
# Paso 01: identificar los parámetros
a = 1
gamma = 2
omega = 10

# Paso 02: calcular los autovalores
characteristical_polynomial = np.poly1d([a, gamma, omega])
roots = np.roots(characteristical_polynomial)
p1 = roots[0]
p2 = roots[1]
print("P1 = {:.2f} y P2 = {:.2f}".format(p1,p2))
#  Paso Extra, función para crear el sistema lineal:
x_1 = lambda t: np.exp(-t) * np.cos(3 * t)
x_2 = lambda t: np.exp(-t) * np.sin(3 * t)
x_1_derivative = lambda t: -np.exp(-t) * np.cos(3 * t) - 3 * np.exp(-t) * np.sin(3 * t)
x_2_derivative = lambda t: -np.exp(-t) * np.sin(3 * t) + 3 * np.exp(-t) * np.cos(3 * t)

# Paso 03: calcular los coeficientes C1 y C2 a partir de las condiciones iniciales
A = np.array([[x_1(0), x_2(0)],[x_1_derivative(0), x_2_derivative(0)]])
b = np.array([[1], [1]])
x_solution = np.linalg.solve(A, b)
C1 = x_solution[0][0]
C2 = x_solution[1][0]

print("C1 = {:.2f} y C2 = {:.2f}".format(C1,C2))

# Paso 04: construir función que representa la solución analítica
def underdamped(lambda_0, c_1, c_2, t):
    """Returns analytic solution x(t)"""
    return np.exp(lambda_0.real * t) * (c_1 * np.cos(lambda_0.imag * t) + c_2 * np.sin(lambda_0.imag * t))
# Paso 05: crear vector de tiempo
t0 = 0
tf = 10
dt = 0.02
vect_time_critical = np.arange(t0, tf + dt, dt)
# Paso 06: solucionar
underdamped_analytic_function = underdamped(p1, C1, C2, vect_time_critical)

# Paso 07: Graficar y carpinteria
fig, ax = plt.subplots(1, 1, figsize=(8,6), dpi=100)
ax.plot(vect_time_critical, underdamped_analytic_function, 'g--', label="Solución Análitica")
ax.grid()
ax.set_title("Sistema Sub Amortiguado")
ax.legend()
ax.set_xlim([0, tf])
```
