### Introducción

<div align='justify'>El modelo de vaciado de tanque, al igual que el modelo compartimental de tanques interconectados, proviene del Teorema de Transporte de Reynolds empleado en la conservación de la masa. </div><br>

1. ¿Qué dice el TTR para la conservación de la masa?
<br>
$$
\frac{D M_{\text {sis }}}{D t}=\frac{\partial}{\partial t} \int_{\mathrm{vc}} \rho  d \forall+\int_{\mathrm{sc}} \rho  \mathbf{V} \cdot \hat{\mathbf{n}} d A \tag{1}
$$
<br>

1. Podemos hacer tres simplificaciones:
<br>
$$
\frac{D M_{\text {sis }}}{D t} = 0
$$
<br>
$$
\frac{\partial}{\partial t} \int_{\mathrm{vc}} \rho  d \forall = \frac{dm_{vc}}{dt}
$$
<br>
$$
\int_{\mathrm{sc}} \rho  \mathbf{V} \cdot \hat{\mathbf{n}} d A = \sum \dot m_{salen} - \sum \dot m_{entran}
$$
<br>
1. Actualizamos la ecuación 1:
<br>
$$
    0 = \frac{dm_{vc}}{dt} + \sum \dot m_{salen} - \sum \dot m_{entran}
$$
<br>
Y si despejamos el término $\frac{dm_{vc}}{dt}$:
<br><br>
$$
    \frac{dm_{vc}}{dt} =  \sum \dot m_{entran} - \sum \dot m_{salen} \tag{2}
$$
<br>

<div align='justify'>En pocas palabras, la ecuación 2 indica que "El cambio en el tiempo de la masa contenida en el Volumen de Control (VC) es igual a la diferencia entre el total de flujo másico que entró y el total de flujo másico que salió". </div><br>

<b> Paso 03: Crear una función en donde se resuelva la ecuación 4 de vaciado de tanque. La función debe aceptar un vector de tiempo, el paso temporal y los parámetros del modelo. </b>

```python
# Problema 0: definir función para evaluar la pendiente
def derivative_function(h, g, a, at, c):
    """Funcion derivada dh/dt """
    return -a * c * np.sqrt(np.abs(2 * g * h)) / at
# Problema 1: definir función para usar Euler, esquema dei ntegración explicito de primer orden
def euler_forward(vect_time, dt, info):
    """Esquema de Integración de Primer Orden"""
    # Asignar valores
    h0, g, a, at, c = info
    # Funcion Solución
    numerical_height = np.zeros(len(vect_time))
    # Condicion inicial
    numerical_height[0] = h0
    # Solución
    for i in range(len(vect_time) - 1):
        numerical_height[i + 1] = numerical_height[i] + derivative_function(numerical_height[i], g, a, at, c) * dt
    # Retorne lo necesario    
    return numerical_height
```

```python
# Problema 0: definir función para evaluar la pendiente
def derivative_function_no_cd(h, g, a, at):
    """Funcion derivada dh/dt """
    return -a * np.sqrt(np.abs(2 * g * h)) / at
# Problema 1: definir función para usar Euler, esquema dei ntegración explicito de primer orden
def euler_forward_no_cd(vect_time, dt, info):
    """Esquema de Integración de Primer Orden"""
    # Asignar valores
    h0, g, a, at = info
    # Funcion Solución
    numerical_height = np.zeros(len(vect_time))
    # Condicion inicial
    numerical_height[0] = h0
    # Solución
    for i in range(len(vect_time) - 1):
        numerical_height[i + 1] = numerical_height[i] + derivative_function_no_cd(numerical_height[i], g, a, at) * dt
    # Retorne lo necesario    
    return numerical_height
```

<b> Modelo Legendario </b>
</div><br>


$$
    h \left(g+ \frac{d^2h}{dt^2} \right) = \frac{1}{2}\left(\frac{dh}{dt} \right)^2 \left[ \left(\frac{A}{a}\right)^2-1 \right]
$$

<div class="alert alert-block alert-success">
<b> Modelo semi-legendario </b>
</div><br>

¿Cómo se puede resovler este modelo? Una opción es suponer que $g >> \frac{d^2h}{dt^2}$, y por tanto, se desprecia la derivada de segundo orden:

$$
    h g = \frac{1}{2}\left(\frac{dh}{dt} \right)^2 \left[ \left(\frac{A}{a}\right)^2-1 \right]
$$

Si se despeja $\frac{dh}{dt}$ se obtiene la siguiente ecuación:

$$
  \frac{dh}{dt} = -\left\{ \frac{2hg}{\left[ \left(\frac{A}{a}\right)^2-1 \right]}\right\}^{1/2}
$$




Si aún así no está convencido con los anteriores modelos, se puede resolver el modelo completo:

$$
    h \left(g+ \frac{d^2h}{dt^2} \right) = \frac{1}{2}\left(\frac{dh}{dt} \right)^2 \left[ \left(\frac{A}{a}\right)^2-1 \right]
$$

# Para realizar el modelo legendario:
Qué necesitamos? Lo mejor es llevar esa Edo de segundo orden no lineal a un sistema de EDOS de primer orden que aunque siga sin ser lineal, puede ser mucho más fácil de trabajar, para ello realizo la sustitución clasica.
$$
y_1 =h \xrightarrow{Derivando} y_1' = y_2, \quad y_2 = h' \xrightarrow{Derivando} y_2' = h''
$$
Aplicamos nuestra susticuion y buscamos solucionar
$$
\begin{cases}
y_1' = y_2 \\
y_1(g+y_2') = \dfrac{1}{2}y_2^2\left[\left( \dfrac{A}{a}^2-1 \right)\right] \Longrightarrow y_2' =\dfrac{y_2^2\left[\left( \dfrac{A}{a}^2-1 \right)\right]}{2y_1}-g
\end{cases}
$$

```python
# Semi Legendario


def semi_legendary_model_dh(h, g, at, a):
    return -np.sqrt(2 * h * g / ((at / a)**2 - 1))
# Legendario


def legendary_model_second_dh(y_1, y_2, g, at, a):
    """Euler hacía adelante de la segunda derivada, resulta en que conocemos la solución numerica de la primera derivada"""
    fraccion = (at / a)**2 - 1
    return (y_2**2 / (2 * y_1) * (fraccion))  - g
    
# Solucion Semi Legendario
semi_legendary_solution = np.zeros(len(vect_time))
semi_legendary_solution[0] = info[0]
for i in range(len(vect_time) - 1):
    semi_legendary_solution[i + 1] = semi_legendary_solution[i] + semi_legendary_model_dh(semi_legendary_solution[i], g, at, a) * dt

# Solcion legendario
# Legendary_solution se puede ver como y_1 y legendary_derivative se puede ver como y_2, posición y velocidad respectivamente
vect_time_leg = np.arange(t0,tf + dt, dt)
legendary_derivative = np.zeros(len(vect_time_leg))
legendary_solution = np.zeros(len(vect_time_leg))
legendary_solution[0] = info[0]
for i in range(len(vect_time_leg) - 1):
    legendary_derivative[i + 1] = legendary_derivative[i] + legendary_model_second_dh(legendary_solution[i], legendary_derivative[i], g, at, a) * dt 
    legendary_solution[i + 1] = legendary_solution[i] + legendary_derivative[i+1] * dt

    
fig, ax = plt.subplots(1, 1, figsize=(6,4), dpi=150)
ax.plot(experimental_time, experimental_height , '-oy', label=fr"Datos Reportados")
ax.plot(vect_time, numerical_height, '-.c', label=fr"Forma Numerica; $C_d = {info[-1]}$")
ax.plot(vect_time, analytical_height, 'r--', label=fr"Forma análitica; $C_d = {info[-1]}$")
ax.plot(vect_time, semi_legendary_solution, 'k--', label=fr"Forma semi_legendaria")
#ax.plot(vect_time_leg, legendary_solution, 'm.-', label=fr"Forma legendaria")

ax.grid()
ax.legend()
ax.set_xlabel("t[s]")
ax.set_ylabel("Altura [cm]")
ax.set_xlim([0, 60])
ax.set_ylim(0, None)
```

<div class="alert alert-block alert-danger">
<b> Bonus. Existe una manera más exacta de hallar el coeficiente de descarga: haciendo una regresión lineal de un modelo potencial</b>
</div><br>  

Pero... ¿Cómo se hace una regresión lineal de un modelo potencial? Aquí va:

Podemos plantear una relación del tipo pontencial entre la altura $h$ y el caudal $\dot \forall$

$$
\dot \forall(h) = \delta\,h^{n}
$$

donde $\delta$ y $n$ son el coeficiente de proporcionalidad y la potencia del modelo, respectivamente.

Para poder determinar los valores de $\delta$ y $n$ dado un conjunto de valores experimentales para $h$, podemos hacer regresión lineal haciendo un pequeño ajuste del modelo.  Para tal fin se toman los logaritmos naturales (o cualquier otro logaritmo) de cada lado del modelo de potencia, con lo que se obtiene:

\begin{equation}
\begin{aligned}
\dot \forall &= \delta\,h^{n} \\
\log\left(\dot \forall\right) &= \log\left(\delta\,h^{n}\right)\\
Y &= \log{\delta} + \log{h^{n}}\\
Y &= \log{\delta} + n\,\log{h}\\
\end{aligned}
\end{equation}

Si la ponemos de la siguiente forma:
$$
Y = n\,\log{h} + \log{\delta}  \\
$$


¿No se les parece a algo muy conocido? ¡Tiene la misma forma que una función lineal!

$$
Y = M X + B
$$


Así pues, encontrar el coeficiente $\delta$ y el exponente $n$, es simplemente un proceso de determinar el intersecto con el eje $Y$ y la pendiente de la función.

Supongamos que tenemos datos experimentales de caudal:

```python

# Promedio y desviación estándar
mu, sigma = 0, 500
# Números "aleatorios"
s = np.random.normal(mu, sigma, len(analytical_height))
# Potencia del modelo
n = 0.5
# Coeficiente de proporcionalidad
delta = 0.86*a*np.sqrt(2*g)
# Ejecutamos el modelo potencial
QTest = delta * np.power(analytical_height, n) + s

```

```python
plt.figure(1, figsize=(8,5))
plt.plot(analytical_height, QTest)
plt.xlabel("h [mm]")
plt.ylabel("Q [mm3/s]")
plt.title("Medición experimental de Caudal vs altura del tanque")
plt.grid()
```

```python


# Aplicamos logaritmos natural
Y = np.log(QTest[1:])
X = np.log(analytical_height[1:])

# Juntamos un vector de X y de 1
A = np.vstack([X, np.ones(len(X))]).T

# Encuentrme M y B que minimice la función de costo
M, B = np.linalg.lstsq(A, Y, rcond=None)[0]

print(M)
print(B)

# ¿Cómo encuentro delta?
delta = np.exp(B)
print(delta)

# ¿Cuánto vale el coeficiente de descarga?
CNew = delta/(a * np.sqrt(2 * g))
print(CNew)

```
