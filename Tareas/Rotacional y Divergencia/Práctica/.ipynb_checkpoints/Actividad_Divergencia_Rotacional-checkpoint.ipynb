5. $ f(x,y) = \left(x^3+4x^2y, -5xy^2-y^3\right) \quad \left(0,1\right) \quad \left(1,2\right) \quad \left(1,-1\right) $
6. $ f(x,y) = \left(\cos(x^3) + 4\sin(x^2)\cos(y), -5x^2y^3+2x^3y^2\right) \quad \left(0,0\right) \quad \left(\pi,0\right) \quad \left(0,\pi\right) $

# Resultado de Gradientes
# Divergencias evaluadas en puntos específicos

---
$
f(x,y) = (x^3,-y^3)
$

$
\nabla \cdot f(x,y) = 3x^2 - 3y^2
$


---
$
f(x,y) = (-x^3,y^3)
$

$
\nabla \cdot f(x,y) = -3x^2 + 3y^2
$

---
$
f(x,y) = (x^2y^2,\sin(xy))
$

$
\nabla \cdot f(x,y) = 2xy^2 + x\cos(xy)
$

---
$
f(x,y) = (\sin(x)\cos(y), -\cos(y^2))
$

$
\nabla \cdot f(x,y) = \cos(x)\cos(y) + 2y\sin(y^2)
$

---

$
f(x,y) = (x^3+4x^2y,-5xy^2-y^3)
$

$
\nabla \cdot f(x,y) = 3x^2 - 2xy - 3y^2
$
$
---
$
f(x,y) =
(\cos(x^3) + 4\sin(x^2)\cos(y),
-5x^2y^3 + 2x^3y^2)
$

$
\nabla \cdot f(x,y) =
-3x^2\sin(x^3)
+ 8x\cos(x^2)\cos(y)
- 15x^2y^2
+ 4x^3y
$

```python

# 1. Definimos los ejes (vectores 1D)
range_val = np.linspace(-20, 20, 50)

# 2. CREAMOS LA MALLA
x = np.outer(np.linspace(-3, 3, 100), np.ones(100))
y = x.copy().T
# 3. Calculamos Z usando las matrices
z = 3 * x**2 - 3 * y**2

# 4. Graficamos (usando una proyección 3D)
fig, ax = plt.figure(figsize=(10, 8)), plt.axes(projection='3d') # Importante definir que es 3D
ax.plot_surface(x, y, z, cmap='inferno')
ax.view_init(elev=45, azim=45)
ax.set_title(r"1. $\nabla \cdot f(x,y) = 3x^2-3y^2$", fontsize=16, color='navy', pad=20)
fig.savefig('Imagenes/Ejercicio_1_Divergencia.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python

# 1. Definimos los ejes (vectores 1D)
range_val = np.linspace(-20, 20, 50)

# 2. CREAMOS LA MALLA
x = np.outer(np.linspace(-3, 3, 100), np.ones(100))
y = x.copy().T
# 3. Calculamos Z usando las matrices
z = 2 * x * y**2 + np.cos(x * y) * x

# 4. Graficamos (usando una proyección 3D)
fig, ax = plt.figure(figsize=(10, 8)), plt.axes(projection='3d') # Importante definir que es 3D
ax.plot_surface(x, y, z, cmap='inferno')
ax.view_init(elev=45, azim=135)
ax.set_title(r"3. $\nabla \cdot f(x,y) = 2xy^2+cos(xy)x$", fontsize=16, color='navy', pad=20)
fig.savefig('Imagenes/Ejercicio_3_Divergencia.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 1. Definimos los ejes (vectores 1D)
range_val = np.linspace(-20, 20, 50)

# 2. CREAMOS LA MALLA
x = np.outer(np.linspace(-5, 5, 50), np.ones(50))
y = x.copy().T

# 3. Calculamos Z usando las matrices
z = 3 * x**2 - 2 * x * y - 3 * y**2
# 4. Graficamos (usando una proyección 3D)
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d') # Importante definir que es 3D
ax.plot_surface(x, y, z, cmap='inferno')
ax.set_title(r"5. $\nabla \cdot f(x,y) = 3x^2 - 2xy - 3y^2$", fontsize=16, color='navy', pad=20)
fig.savefig('Imagenes/Ejercicio_5_Divergencia.png', dpi=150, bbox_inches='tight')

plt.show()
```

# Rotacionales Resultantes

1. $
\nabla \times f(x,y) = (0,0,0)
$

2. $
\nabla \times f(x,y) = (0,0,0)
$

3. $
\nabla \times f(x,y) =
\left(0,0,
y\cos(xy) - 2x^2y
\right)
$

4. $
\nabla \times f(x,y) =
\left(0,0,
\sin(x)\sin(y)
\right)
$

5. $
\nabla \times f(x,y) =
\left(0,0,
-5y^2 - 4x^2
\right)
$

6. $
\nabla \times f(x,y) =
\left(0,0,
4\sin(x^2)\sin(y)
-10xy^3
+6x^2y^2
\right)
$

```python
# 1D arrays
x = np.arange(-5,5,0.2)
y = np.arange(-5,5,0.2)

# Meshgrid
X,Y = np.meshgrid(x,y)

# rotacional (componente k)
rot = Y*np.cos(X*Y) - 2*X**2*Y

fig = plt.figure(figsize=(10,10))

plt.contourf(X,Y,rot, levels=40, cmap="inferno")
plt.colorbar(label="rotacional")

plt.title(r'3. $\nabla \times f(x,y) = \left(0,0,y\cos(xy)-2x^2y\right)$')

plt.grid()

#fig.savefig('Imagenes/Ejercicio_3_Rotacional.png', dpi=150, bbox_inches='tight')

plt.show()
```

```python
# dominio
x = np.arange(-5,5,0.2)
y = np.arange(-5,5,0.2)

X,Y = np.meshgrid(x,y)

# rotacional (campo escalar)
Z = Y*np.cos(X*Y) - 2*X**2*Y

# grafica 3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X,Y,Z)

ax.set_title(
r"3. $z = y\cos(xy) - 2x^2y$",
fontsize=16,
pad=20
)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("rotacional")

plt.show()
```

```python

# 1D arrays
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
fx = -3*X**2
fy = 3*Y**2

# Depict illustration
fig = plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,fx,fy, density=1.4, linewidth=None, color='#C7C481')
plt.title(r'2. $ \nabla\cdot f(x,y) = \left(-3x^2 , 3y^2 \right) $')

# Show plot with grid
plt.grid()
fig.savefig('Ejercicio_2_Vectorial.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 1D arrays
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
fx = 2 * X * Y**2 - Y*np.cos(X * Y)
fy = 2 * X**2 * Y - X*np.cos(X * Y)

# Depict illustration
fig = plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,fx,fy, density=1.4, linewidth=None, color='#C45CC4')
plt.title(r'3. $ \nabla\cdot f(x,y) = \left(2xy^2-ycos(xy) , 2x^2y-xcos(xy) \right) $')

# Show plot with grid
plt.grid()
fig.savefig('Ejercicio_3_Vectorial.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 1D arrays
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
fx = np.cos(X) * np.cos(Y)
fy = 2 * Y * np.sin(Y**2) - np.sin(X) * np.sin(Y)

# Depict illustration
fig = plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,fx,fy, density=1.4, linewidth=None, color='#C45C5C')
plt.title(r'4. $ \nabla\cdot f(x,y) = \left(cos(x)cos(y), 2ysin(y^2)-sin(x)sin(y)) \right)$')

# Show plot with grid
plt.grid()
fig.savefig('Ejercicio_4_Vectorial.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 1D arrays
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
fx = 3 * X**2 + 8 * X * Y - 5 * Y**2
fy = 4 * X**2 - 10 * X * Y - 3 * Y**2

# Depict illustration
fig = plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,fx,fy, density=1.4, linewidth=None, color='#73C45C')
plt.title(r'5. $ \nabla\cdot f(x,y) = \left(3x^2+8xy-5y^2, 4x^2 -10xy - 3y^2\right)$')

# Show plot with grid
plt.grid()
fig.savefig('Ejercicio_5_Vectorial.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 1D arrays
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
fx = -3 * X**2 * np.sin(X**3) + 8 * X * np.cos(X**2) * np.cos(Y) -5 * Y**2
fy = -4 * np.sin(X**2) * np.sin(Y) - 10 * X * Y + 3 * Y**2

# Depict illustration
fig = plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,fx,fy, density=1.4, linewidth=None, color='#722F37')
plt.title(r'6. $ \nabla\cdot f(x,y) = \left(-3x^2sin(x^3) + 8xcos(x^2)cos(y) - 5y^2, -4sin(x^2)sin(y)-10xy + 3y^2\right)$')

# Show plot with grid
plt.grid()
fig.savefig('Ejercicio_6_Vectorial.png', dpi=150, bbox_inches='tight')
plt.show()
```
